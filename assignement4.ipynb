```python
import spacy
from sklearn.metrics import classification_report, accuracy_score

# Download the spaCy model
import os
os.system("python -m spacy download en_core_web_sm")

nlp = spacy.load("en_core_web_sm")
```

```python
texts = [
    "Apple CEO Tim Cook announced a new product in California on Monday.",
    "Elon Musk acquired Twitter for $44 billion in 2022."
]
```

```python
predicted_entities = []

for text in texts:
    doc = nlp(text)
    for ent in doc.ents:
        predicted_entities.append((ent.text, ent.label_))

print(predicted_entities)
```

```python
true_entities = [
    ("Apple", "ORG"),
    ("Tim Cook", "PERSON"),
    ("California", "GPE"),
    ("Monday", "DATE"),
    ("Elon Musk", "PERSON"),
    ("Twitter", "ORG"),
    ("$44 billion", "MONEY"),
    ("2022", "DATE")
]
```

```python
y_true = [label for _, label in true_entities]
y_pred = [label for _, label in predicted_entities]
```

```python
print("Accuracy:", accuracy_score(y_true, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_true, y_pred))
```
