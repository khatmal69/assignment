```python
# Core NLP library
import nltk

# Built-in dataset from NLTK
from nltk.corpus import gutenberg

# Tokenization
from nltk.tokenize import word_tokenize

# Text vectorization techniques
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Word Embedding model
from gensim.models import Word2Vec

# Numerical operations
import numpy as np
```

```python
# Download tokenizer and dataset
nltk.download('punkt')
nltk.download('gutenberg')
```

```python
# Load text data from Gutenberg corpus
# Using Jane Austen's novel "Emma"
text_data = gutenberg.raw('austen-emma.txt')

# Display sample text
print("Sample Dataset Text:\n")
print(text_data[:400])
```

```python
# Convert text into sentences
# Many vectorizers work better sentence-wise
sentences = text_data.split('.')

# Remove empty sentences
sentences = [sentence.strip() for sentence in sentences if sentence.strip()]

print("Number of sentences:", len(sentences))
print("\nSample sentence:\n", sentences[0])
```

```python
# CountVectorizer converts text into a matrix of word counts
# Each row = sentence, each column = word

count_vectorizer = CountVectorizer(stop_words='english')

# Fit and transform sentences
bow_count = count_vectorizer.fit_transform(sentences)

# Convert to array for easy viewing
bow_count_array = bow_count.toarray()

print("Bag of Words - Count Occurrence Matrix Shape:")
print(bow_count_array.shape)

print("\nSample BoW Count Vector (first sentence):")
print(bow_count_array[0])
```

```python
# Normalize BoW counts so that values lie between 0 and 1
# This removes bias toward longer sentences

normalized_bow = bow_count_array / np.sum(bow_count_array, axis=1, keepdims=True)

print("Normalized Bag of Words (first sentence):")
print(normalized_bow[0])
```

```python
# TfidfVectorizer assigns importance to words
# Common words get lower weight, rare words get higher weight

tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform sentences
tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)

# Convert to array
tfidf_array = tfidf_matrix.toarray()

print("TF-IDF Matrix Shape:")
print(tfidf_array.shape)

print("\nSample TF-IDF Vector (first sentence):")
print(tfidf_array[0])
```

```python
# Word2Vec requires tokenized sentences
tokenized_sentences = [word_tokenize(sentence.lower()) for sentence in sentences]

print("Sample Tokenized Sentence:\n", tokenized_sentences[0])
```

```python
# Word2Vec learns semantic relationships between words
# vector_size = embedding dimension
# window = context window size
# min_count = ignores rare words

word2vec_model = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4
)
```

```python
# Get embedding vector for a word
word = "emma"

if word in word2vec_model.wv:
    embedding = word2vec_model.wv[word]
    print(f"Word2Vec Embedding for '{word}':\n")
    print(embedding)
    print("\nEmbedding Vector Length:", len(embedding))
else:
    print("Word not found in vocabulary")
```
