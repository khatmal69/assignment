```python
import nltk

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
text = """Natural Language Processing enables computers to understand human language.
It is widely used in chatbots, search engines, and sentiment analysis."""
```

```python
# Sentence Tokenization
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)
print(sentences)
```

```python
# Word Tokenization
from nltk.tokenize import word_tokenize

words = word_tokenize(text)
print(words)
```

```python
# Converting to Lowercase
words_lower = [w.lower() for w in words]
print(words_lower)
```

```python
# Removing Stopwords
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words_lower if w.isalpha() and w not in stop_words]

print(filtered_words)
```

```python
# Stemming
from nltk.stem import PorterStemmer

ps = PorterStemmer()
stemmed_words = [ps.stem(w) for w in filtered_words]

print(stemmed_words)
```

```python
# Lematization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered_words]
print(lemmatized_words)
```

```python
# POS Tagging on Tokens
from nltk import pos_tag

pos_tags = pos_tag(words)
print(pos_tags)
```

```python
# Word frequency distribution
from nltk.probability import FreqDist

freq_dist = FreqDist(filtered_words)
print(freq_dist.most_common(10))
```

```python
#Text Length Statistics
print("Total sentences:", len(sentences))
print("Total words:", len(words))
print("Filtered words:", len(filtered_words))
```

```python
# Combined function which can directly work on the i/p TEXT
def preprocess_text(text):
    words = word_tokenize(text.lower())
    stop_words = set(stopwords.words('english'))
    filtered = [w for w in words if w.isalpha() and w not in stop_words]
    lemmatizer = WordNetLemmatizer()
    lemmatized = [lemmatizer.lemmatize(w) for w in filtered]
    pos = pos_tag(lemmatized)
    return lemmatized, pos
```

```python
new_text = """Machine learning is a subset of artificial intelligence that focuses on building systems that learn from data.
It is used in various applications such as recommendation systems, fraud detection, and image recognition."""

tokens, pos_tags = preprocess_text(new_text)
print("Tokens:", tokens)
print("POS Tags:", pos_tags)
```
