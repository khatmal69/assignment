```python
from nltk.corpus import wordnet as wn
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
```

```python
def perform_wsd(sentence, ambiguous_word):
    tokens = word_tokenize(sentence)
    sense = lesk(tokens, ambiguous_word)
    
    if sense:
        print("Sentence:", sentence)
        print("Ambiguous Word:", ambiguous_word)
        print("Selected Sense:", sense.name())
        print("Definition:", sense.definition())
        print("Examples:", sense.examples())
        print("-" * 60)
    else:
        print("No sense found.")
```

```python
sentence1 = "I deposited money in the bank"
sentence2 = "He sat on the bank of the river"

perform_wsd(sentence1, "bank")
perform_wsd(sentence2, "bank")
```

```python
sentence3 = "He hit the ball with a bat"
sentence4 = "A bat flew in the dark"

perform_wsd(sentence3, "bat")
perform_wsd(sentence4, "bat")
```

```python
sentence5 = "The workers built a new plant for manufacturing cars"
sentence6 = "She watered the plant in the garden"

perform_wsd(sentence5, "plant")
perform_wsd(sentence6, "plant")
```

```python
def show_all_senses(word):
    print("All possible senses of:", word)
    for synset in wn.synsets(word):
        print(synset.name(), ":", synset.definition())
    print("-" * 60)

show_all_senses("bank")
```
