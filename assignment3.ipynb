```python
# Core NLP library
import nltk

# Dataset from NLTK
from nltk.corpus import gutenberg, stopwords

# Tokenization
from nltk.tokenize import word_tokenize

# Lemmatization
from nltk.stem import WordNetLemmatizer

# Text processing
import re
import string

# Data handling
import pandas as pd

# Encoding and vectorization
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Download required datasets and tools
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
nltk.download('gutenberg')
```

```python
# Load text from Gutenberg corpus
text_data = gutenberg.raw('austen-emma.txt')

# Convert text into sentences
sentences = text_data.split('.')

# Remove empty sentences
sentences = [sentence.strip() for sentence in sentences if sentence.strip()]

print("Total Sentences:", len(sentences))
print("\nSample Sentence:\n", sentences[0])
```

```python
# Function for text cleaning
def clean_text(text):
    """
    This function:
    - Converts text to lowercase
    - Removes numbers
    - Removes punctuation
    - Removes extra spaces
    """
    text = text.lower()
    text = re.sub(r'\d+', '', text)  # remove numbers
    text = text.translate(str.maketrans('', '', string.punctuation))  # remove punctuation
    text = re.sub(r'\s+', ' ', text)  # remove extra spaces
    return text.strip()

# Apply text cleaning
cleaned_sentences = [clean_text(sentence) for sentence in sentences]

print("Cleaned Text Sample:\n", cleaned_sentences[0])
```

```python
# Initialize WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize each sentence
lemmatized_sentences = [
    " ".join([lemmatizer.lemmatize(word) for word in word_tokenize(sentence)])
    for sentence in cleaned_sentences
]

print("Lemmatized Text Sample:\n", lemmatized_sentences[0])
```

```python
# Load English stopwords
stop_words = set(stopwords.words('english'))

# Remove stopwords
stopword_removed_sentences = [
    " ".join([word for word in word_tokenize(sentence) if word not in stop_words])
    for sentence in lemmatized_sentences
]

print("Stopword Removed Text Sample:\n", stopword_removed_sentences[0])
```

```python
# Create simple labels for demonstration
# Example: classify sentences as Short or Long
labels = ['Long' if len(sentence.split()) > 8 else 'Short'
          for sentence in stopword_removed_sentences]

# Apply Label Encoding
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

print("Original Labels:", labels[:10])
print("Encoded Labels:", encoded_labels[:10])
```

```python
# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Create TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(stopword_removed_sentences)

print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
```

```python
# Convert TF-IDF matrix to DataFrame
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf_vectorizer.get_feature_names_out()
)

# Create final dataset with labels
final_df = pd.concat([
    pd.DataFrame(stopword_removed_sentences, columns=['Processed_Text']),
    pd.DataFrame(encoded_labels, columns=['Encoded_Label'])
], axis=1)

# Save outputs to CSV files
tfidf_df.to_csv("tfidf_features.csv", index=False)
final_df.to_csv("processed_text_with_labels.csv", index=False)

print("Files saved successfully:")
print("1. tfidf_features.csv")
print("2. processed_text_with_labels.csv")
```
