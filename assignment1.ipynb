```python
# Import core NLTK package
import nltk

# Import Gutenberg corpus (built-in dataset)
from nltk.corpus import gutenberg

# Import different tokenizers
# Each tokenizer splits text in a different way
from nltk.tokenize import (
    word_tokenize,            # Standard word tokenizer
    WhitespaceTokenizer,      # Tokenizes based on spaces
    TreebankWordTokenizer,    # Tokenizer used in Penn Treebank
    TweetTokenizer,           # Designed for social media text
    MWETokenizer              # Handles multi-word expressions
)

# Import stemming algorithms
from nltk.stem import PorterStemmer, SnowballStemmer

# Import lemmatizer
from nltk.stem import WordNetLemmatizer

# Used to handle punctuation symbols
import string
```

```python
# Download tokenizer models
nltk.download('punkt')

# Download WordNet dictionary for lemmatization
nltk.download('wordnet')
nltk.download('omw-1.4')

# Download Gutenberg corpus dataset
nltk.download('gutenberg')
```

```python
# Load text from Gutenberg corpus
# We are using "Emma" novel by Jane Austen
text_data = gutenberg.raw('austen-emma.txt')

# Display only a small portion of text
print("Sample Dataset Text:\n")
print(text_data[:500])
```

## TOKANIZATION


```python
# WhitespaceTokenizer splits text wherever there is a space
# It is the simplest form of tokenization

whitespace_tokenizer = WhitespaceTokenizer()
whitespace_tokens = whitespace_tokenizer.tokenize(text_data)

print("Whitespace Tokenization Output (first 50 tokens):\n")
print(whitespace_tokens[:50])
```

```python
# This method removes punctuation marks from words
# It splits text using spaces and then strips punctuation

punctuation_tokens = [
    word.strip(string.punctuation)
    for word in text_data.split()
    if word.strip(string.punctuation)
]

print("Punctuation-based Tokenization Output (first 50 tokens):\n")
print(punctuation_tokens[:50])
```

```python
# TreebankWordTokenizer follows grammar rules
# It handles contractions like "don't" → "do" + "n't"

treebank_tokenizer = TreebankWordTokenizer()
treebank_tokens = treebank_tokenizer.tokenize(text_data)

print("Treebank Tokenization Output (first 50 tokens):\n")
print(treebank_tokens[:50])
```

```python
# TweetTokenizer is optimized for social media text
# It preserves emojis, hashtags, and mentions

tweet_tokenizer = TweetTokenizer()
tweet_tokens = tweet_tokenizer.tokenize(text_data)

print("Tweet Tokenization Output (first 50 tokens):\n")
print(tweet_tokens[:50])
```

```python
# MWETokenizer combines frequently occurring word groups
# Example: "natural language processing" → "natural_language_processing"

mwe_tokenizer = MWETokenizer([
    ('natural', 'language', 'processing'),
    ('machine', 'learning'),
    ('artificial', 'intelligence')
], separator='_')

# word_tokenize is used before MWE tokenization
mwe_tokens = mwe_tokenizer.tokenize(word_tokenize(text_data))

print("MWE Tokenization Output (first 50 tokens):\n")
print(mwe_tokens[:50])
```

## STEMMING

```python
# PorterStemmer is a rule-based stemming algorithm
# It removes suffixes to reduce words to root form

porter = PorterStemmer()
porter_stems = [porter.stem(word) for word in word_tokenize(text_data)]

print("Porter Stemmer Output (first 50 words):\n")
print(porter_stems[:50])
```

```python
# SnowballStemmer is an improved version of Porter
# It provides better accuracy and supports multiple languages

snowball = SnowballStemmer("english")
snowball_stems = [snowball.stem(word) for word in word_tokenize(text_data)]

print("Snowball Stemmer Output (first 50 words):\n")
print(snowball_stems[:50])
```

## LEMMATIZATION

```python
# WordNetLemmatizer reduces words to their dictionary form
# Example: "running" → "run"

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in word_tokenize(text_data)]

print("Lemmatization Output (first 50 words):\n")
print(lemmatized_words[:50])
```
