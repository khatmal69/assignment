**🔹 What is Federated Learning?**

Federated Learning is a distributed machine learning approach where clients train models locally and only model weights are shared, not raw data.

**🔹 How it works:**

1. Server sends global model to clients

2. Clients train locally on private data

3. Clients send updated weights

4. Server performs Federated Averaging (FedAvg)

5. Updated global model is redistributed

**🔹 Advantages:**

Data privacy ✔

Reduced data transfer ✔

Secure learning ✔

   
   **FEDERATED LEARNING DIAGRAM**

        Client 3                         ▼
        ┌───────────────┐         Global Model
        │ Train Model   │
        └───────────────┘

```python
import numpy as np
import tensorflow as tf

def model():
    return tf.keras.Sequential([tf.keras.layers.Dense(10, activation='softmax', input_shape=(784,))])

def client_train(m, x, y):
    m.compile(optimizer='sgd', loss='sparse_categorical_crossentropy')
    m.fit(x, y, epochs=1, verbose=0)
    return m.get_weights()

clients_x = [np.random.rand(100,784) for _ in range(3)]
clients_y = [np.random.randint(0,10,100) for _ in range(3)]

global_model = model()
for r in range(3):
    weights = []
    for i in range(3):
        local_model = model()
        local_model.set_weights(global_model.get_weights())
        weights.append(client_train(local_model, clients_x[i], clients_y[i]))
    global_model.set_weights([np.mean(w, axis=0) for w in zip(*weights)])
    print("Round", r+1, "completed")
```
